```python
Numerical = ["Age", "Fare"]
```

```python
titanic_normalized = titanic.copy()
titanic_normalized[Numerical] = (titanic[Numerical]-titanic[Numerical].mean())/titanic[Numerical].std()
```

```python
mean_y_no_yes = titanic_normalized.groupby(titanic['Survived']).mean()
print(mean_y_no_yes)
std_y_no_yes = titanic_normalized.groupby(titanic['Survived']).std()
print(std_y_no_yes)
```

```python
import matplotlib.pyplot as plt

from scipy import stats
from scipy.stats import norm
from scipy.interpolate import interp1d

for j in titanic_normalized[Numerical]:
    plot_yes_x = titanic_normalized.loc[(titanic_normalized['Survived'])==1, j].tolist()
    plot_no_x = titanic_normalized.loc[(titanic_normalized['Survived'])==0, j].tolist()
    plot_yes_x.sort()
    plot_no_x.sort()
    plt.title(j)
    plt.xlabel(j)
    plt.ylabel('probabilty')
    f = stats.norm.pdf(plot_yes_x, mean_y_no_yes[j][1], std_y_no_yes[j][1])
    g = stats.norm.pdf(plot_no_x, mean_y_no_yes[j][0], std_y_no_yes[j][0])
    ff = interp1d(plot_yes_x, f, kind = 'linear')
    gg =  interp1d(plot_no_x, g, kind = 'linear')
    xx = np.linspace(max(plot_yes_x[0], plot_no_x[0]), min(plot_yes_x[-1], plot_no_x[-1]), 1000)
    f_interp1d = ff(xx)
    g_interp1d = gg(xx)
    index = np.argwhere(np.diff(np.sign(f_interp1d-g_interp1d))).flatten()
    plt.plot(plot_yes_x, f, 'red')
    plt.plot(plot_no_x, g, 'green')
    for ind in index:
        print('intersection_point = ',xx[ind])
        plt.plot(xx[ind], f_interp1d[ind], marker = 'o', markerfacecolor = 'black', markeredgecolor = 'black')
    plt.show()
```
